```python
import numpy as np
import torch
```

```python
def plot_state_space(model, inp, ranges = [], dims = [], resolutions = 100):
    batched_data, mesh = create_mesh_state_space(inp, ranges,dims, resolutions)
    outs = model.predict(batched_data)
    
    """
    Params
        inp : (dim)
    """

def create_mesh_state_space(inp, ranges = [], dims = [], resolutions = 100):
    if isinstance(resolutions, int):
        resolutions = np.asarray([resolutions] * len(ranges))
    else:
        resolutions = np.asarray(resolutions)

    mesh_space = [np.linspace(*rng, res) for rng,res in zip(ranges,resolutions)]
    mesh = np.asarray([k.flatten() for k in np.meshgrid(*mesh_space)]).T
    mesh = torch.tensor(mesh)
    
    dim = resolutions.prod()
    inp = torch.tensor(inp, dtype = torch.float32)
    batched_data = inp.repeat(dim,*([1]*inp.dim()))

    for dim in dims:
        batched_data[:, dim] = mesh[:, dim]

    return batched_data, mesh


inp = torch.tensor([0,0,0,0])
ranges = [(-2,2),(-2,2)]
dims = [0,1]

r = plot_state_space(None,inp,ranges,dims)
r.shape
```
